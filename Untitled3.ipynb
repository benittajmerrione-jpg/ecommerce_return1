```python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)

IQR = Q3 - Q1

outliers = df[
    (df["price"] < (Q1 - 1.5*IQR)) |
    (df["price"] > (Q3 + 1.5*IQR))
]

print("Outliers Found =", len(outliers))
```

```python
from scipy.stats import zscore

df["zscore"] = zscore(df["price"])

print(df[df["zscore"].abs() > 3])
```

```python
from scipy.stats import ttest_ind

returned = df[df["is_returned"]==1]["shipping_time_days"]

not_returned = df[df["is_returned"]==0]["shipping_time_days"]

t_stat,p_value = ttest_ind(returned,not_returned)

print("T Statistic =", t_stat)
print("P Value =", p_value)
```

```python
if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```
